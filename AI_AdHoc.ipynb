```
import matplotlib.pyplot as plt
```

```
x=[2,9,3,7]
y=[7,15,5,9]
```

```
plt.plot(x,y,label="railway tracks")
plt.xlabel('Distance')
plt.ylabel('Time')
plt.scatter(x,y,s=200,label='railway station',marker='^',color='red')
plt.grid(color='green')
plt.bar(x,y,label="height of railway station")
plt.legend()
plt.show()
```
